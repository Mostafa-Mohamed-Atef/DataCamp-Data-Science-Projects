```python
parent_themes = pd.read_csv('datasets/parent_themes.csv')
parent_themes.head()
```

```python
lego_joined = df.merge(parent_themes,how='left',left_on = 'parent_theme',right_on = 'name',suffixes=('','_y'))
lego_joined.head()
```

```python
all_sets = len(lego_joined[lego_joined['is_licensed']==True])
all_sets
```

```python
star_wars_licensed = lego_joined[(lego_joined['parent_theme'] == 'Star Wars') & (lego_joined['is_licensed'] == True)]
```

```python
len(star_wars_licensed)
```

```python
ratio = len(star_wars_licensed)/all_sets
the_force = int(ratio*100)
print(the_force)
```

```python
licensed_set = lego_joined[lego_joined['is_licensed']==True]
licensed_set
```

```python
licensed_pivot = licensed_set.pivot_table(index='year',columns='parent_theme',values='set_num',aggfunc='count')
licensed_pivot
```

```python
yearly_max = licensed_pivot.max(axis='columns')
yearly_max
```

```python
filtered_max = licensed_pivot[licensed_pivot['Star Wars']<yearly_max]
filtered_max
```

```python
most_popular_sets = licensed_set.groupby('year')['parent_theme'].agg(pd.Series.mode)
most_popular_sets
```

```python
for year, theme in most_popular_sets.groupby(level = 0):
    if 'Star wars' not in theme.values:
        new_era = year
new_era
```
