```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
sorted_schools = schools.sort_values("average_math", ascending = False)
subsetted = sorted_schools[["school_name", "average_math"]]
best_math_schools = subsetted[subsetted["average_math"] > 640]
best_math_schools
```

```python
schools['total_SAT'] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
schools.head()
```

```python
top_10_schools = schools[["school_name", "total_SAT"]]
top_10_schools = top_10_schools.sort_values("total_SAT", ascending=False).head(10)
top_10_schools
```

```python
borough_stats = schools.groupby('borough')['total_SAT'].agg(num_schools='count', average_SAT='mean', std_SAT='std'
).reset_index()
borough_stats
```

```python
largest_borough = borough_stats.loc[borough_stats['std_SAT'].idxmax()]
largest_borough
```

```python
largest_std_dev = pd.DataFrame({
    'borough': [largest_borough['borough']],
    'num_schools': [largest_borough['num_schools']],
    'average_SAT': [round(largest_borough['average_SAT'], 2)],
    'std_SAT': [round(largest_borough['std_SAT'], 2)]
})
largest_std_dev
```
